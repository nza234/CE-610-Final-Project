```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('surveyresponds.xlsx')


if 'Q9_1' not in data.columns:
    print("Column 'Q9_1' not found in the DataFrame.")
else:
    df_Q9_1 = pd.DataFrame(columns=['Options_Q9_1', 'Frequency', 'Frequency(%)', 'Graphing'])
    Options_Q9_1 = ['Very Low', 'Low', 'Moderate', 'High', 'Very High', 'Uncertain']
    df_Q9_1['Options_Q9_1'] = Options_Q9_1 

    for i in range(0, len(Options_Q9_1)):

        df_Q9_1['Frequency'][i] = data['Q9_1'].astype(str).str.contains(df_Q9_1['Options_Q9_1'][i], regex=False).sum()
        df_Q9_1['Frequency(%)'][i] = round(df_Q9_1['Frequency'][i] / len(data['Q9_1'].dropna()) * 100)

    # Concatenate the percentage information for each option
    df_Q9_1["Graphing"] = df_Q9_1['Frequency(%)'].astype(str) + '% (n=' + df_Q9_1["Frequency"].astype(str) + ')'

    df_Q9_1 = df_Q9_1.dropna(subset=["Options_Q9_1"])
    df_Q9_1.sort_values('Frequency(%)', ascending=True, inplace=True)
    df_Q9_1 = df_Q9_1.reset_index(drop=True)
    print('Sample Size =' + ' ' + str(len(data['Q9_1'].dropna())))
    df_Q9_1

    ax_Q9_1 = df_Q9_1.plot.barh(x='Options_Q9_1', y='Frequency(%)', ylim=100, legend=False, fontsize=14, color='orange',
                                figsize=(10, 5))
    plt.xlabel('Frequency(%)', fontsize=14)
    plt.ylabel('')
    plt.xlim([0, 100])
    top_side = ax_Q9_1.spines["top"]
    top_side.set_visible(False)
    right_side = ax_Q9_1.spines["right"]
    right_side.set_visible(False)

    for i, v in enumerate(df_Q9_1["Frequency(%)"]):
        ax_Q9_1.text(v + 0.5, i - 0.1, str(df_Q9_1["Graphing"][i]), fontsize=11)

    plt.savefig("Q9_1.svg", bbox_inches="tight")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.DataFrame({
    'Group': ['Construction (High)', 'Maintenance (Moderate)', 'Planning (Low)'],
    'Average Percentage Frequency': [79, 29, 50],
})


fig, ax = plt.subplots(figsize=(6, 3.5))
bars = ax.barh(data['Group'], data['Average Percentage Frequency'], color='orange', height=0.4)


for bar, value, group in zip(bars, data['Average Percentage Frequency'], data['Group']):
    ax.text(bar.get_width() + 2, bar.get_y() + bar.get_height() / 2, f'{value}%', 
            va='center', fontsize=10, color='black')


for bar, group in zip(bars, data['Group']):
    ax.text(-2, bar.get_y() + bar.get_height() / 2, f'{group}', 
            va='center', ha='right', fontsize=10, color='black')


ax.set_xlabel('Frequency (%)', fontsize=12)

ax.set_xlim([0, 100]) 


ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_linewidth(0.5)  
ax.spines["left"].set_color('gray')  


ax.tick_params(left=False, labelleft=False)


plt.savefig("Q9.svg", bbox_inches="tight")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def get_frequency_table(data, column_name, options):
    df = pd.DataFrame(columns=[f'Options_{column_name}', 'Frequency', 'Frequency(%)', 'Graphing'])
    df[f'Options_{column_name}'] = options

    for i in range(0, len(options)):
        df['Frequency'][i] = data[column_name].str.contains(df[f'Options_{column_name}'][i], regex=False).sum()
        df['Frequency(%)'][i] = round(df['Frequency'][i] / len(data[column_name].dropna()) * 100)
        df["Graphing"] = df['Frequency(%)'].astype(str) + '% (n=' + df["Frequency"].astype(str) + ')'

    df.sort_values('Frequency(%)', ascending=True, inplace=True)
    df = df.reset_index(drop=True)

    print(f'Sample Size for {column_name} =' + ' ' + str(len(data[column_name].dropna())))
    return df

excel_file_path = 'surveyresponds.xlsx'
data = pd.read_excel(excel_file_path)

columns_and_options = {
    'Q10-3_3': ['Wi-Fi Connectivity to other systems', 'Bluetooth Connectivity to other systems',
                 'Camera', 'Sound Recording', 'Handwriting recording and recognition', 'GIS/GPS capability',
                 'Online field data entry platforms accessibility', 'Offline field data entry platforms accessibility']
}

column = 'Q10-3_3'
result_df = get_frequency_table(data, column, columns_and_options[column])

plt.figure(figsize=(8, 6))
sns.set(font_scale=1.05)  # Adjust font size here
sns.set_style("white")  
ax = sns.barplot(x='Frequency(%)', y=f'Options_{column}', data=result_df, color='orange')

plt.xlabel('Frequency(%)')
plt.ylabel('')

for p in ax.patches:
    width = p.get_width()
    plt.text(width, p.get_y() + p.get_height() / 2, f'{width:.0f}%', ha='left', va='center')

sns.despine(top=True, right=True)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_linewidth(0.5)  
ax.spines["left"].set_color('gray') 

plt.savefig(f'barchart_{column}.svg', format='svg', bbox_inches='tight')
plt.show()
```
